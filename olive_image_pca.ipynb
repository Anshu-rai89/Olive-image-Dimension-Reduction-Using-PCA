```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
olove=datasets.fetch_olivetti_faces()
```

```python
olove.keys()
```

```python
print(olove["data"].shape)
olove['images'].shape
```

```python
fig=plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(olove.images[i],cmap=plt.cm.bone)
    
plt.show()
```

```python
x=olove.data
y=olove.target
pca=PCA()
pca.fit(x)
```

```python
pca.components_.shape
```

```python
k=0
total=0
while total<0.99:
    total+=pca.explained_variance_ratio_[k]
    k=k+1
k
```

```python
pca=PCA(n_components=123)
x_tranformed=pca.fit_transform(x)
x_tranformed.shape
```

```python
x_approx=pca.inverse_transform(x_tranformed)
x_approx.shape
```

```python
x_approx=x_approx.reshape((400,64,64))
```

```python
fig=plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(x_approx[i],cmap=plt.cm.bone)
    
plt.show()
```

```python
#eigen faces
eigenval=pca.components_
eigenval.shape
```

```python
eigenfac=eigenval.reshape((123,64,64))
```

```python
fig=plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(eigenfac[i],cmap=plt.cm.bone)
    
plt.show()
```
